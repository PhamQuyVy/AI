```python
import random

GOAL = [[1,2,3],[4,5,6],[7,8,0]]

LUAT = {
    (0,0): ["DOWN", "RIGHT"],
    (0,1): ["DOWN", "LEFT", "RIGHT"],
    (0,2): ["DOWN", "LEFT"],
    (1,0): ["UP", "DOWN", "RIGHT"],
    (1,1): ["UP", "DOWN", "LEFT", "RIGHT"],
    (1,2): ["UP", "DOWN", "LEFT"],
    (2,0): ["UP", "RIGHT"],
    (2,1): ["UP", "LEFT", "RIGHT"],
    (2,2): ["UP", "LEFT"],
}

DELTA = {"UP":(-1,0), "DOWN":(1,0), "LEFT":(0,-1), "RIGHT":(0,1)}

def display(state):
    for row in state:
        print(" ".join("□" if v == 0 else str(v) for v in row))
    print()
def sensor(state):
    for i in range(3):
        for j in range(3):
            if state[i][j] == 0:
                return (i, j)

def processor(blank):
    r, c = blank
    if r == 0 and c == 0:
        huong = random.choice(["DOWN", "RIGHT"])
    elif r == 0 and c == 1:
        huong = random.choice(["DOWN", "LEFT", "RIGHT"])
    elif r == 0 and c == 2:
        huong = random.choice(["DOWN", "LEFT"])
    elif r == 1 and c == 0:
        huong = random.choice(["UP", "DOWN", "RIGHT"])
    elif r == 1 and c == 1:
        huong = random.choice(["UP", "DOWN", "LEFT", "RIGHT"])
    elif r == 1 and c == 2:
        huong = random.choice(["UP", "DOWN", "LEFT"])
    elif r == 2 and c == 0:
        huong = random.choice(["UP", "RIGHT"])
    elif r == 2 and c == 1:
        huong = random.choice(["UP", "LEFT", "RIGHT"])
    else:
        huong = random.choice(["UP", "LEFT"])
 
    print(f"Blank tai ({r},{c}) -> chon huong: {huong}")
    return huong

def actuator(state, blank, huong):
    r, c = blank
    moi = [hang[:] for hang in state]
 
    if huong == "UP":
        moi[r][c], moi[r-1][c] = moi[r-1][c], moi[r][c]
    elif huong == "DOWN":
        moi[r][c], moi[r+1][c] = moi[r+1][c], moi[r][c]
    elif huong == "LEFT":
        moi[r][c], moi[r][c-1] = moi[r][c-1], moi[r][c]
    elif huong == "RIGHT":
        moi[r][c], moi[r][c+1] = moi[r][c+1], moi[r][c]
 
    print(f"Da di chuyen blank sang {huong}")
    return moi
print("Ma tran ban dau:")
display(state)
state = [row[:] for row in GOAL]
nums = list(range(9))
random.shuffle(nums)

state = [nums[:3],
         nums[3:6],
         nums[6:]]
step = 0
lichsu={}
while True:
    blank = sensor(state)
    move  = processor(blank)
    state = actuator(state, blank, move)
    step += 1
    print(f"Buoc {step}: {move}")
    display(state)

    if state == GOAL:
        print(f"Hoan thanh sau {step} buoc")
        break
    key = str(state)
    lichsu[key] = lichsu.get(key, 0) + 1
    if lichsu[key] >= 30:
        print("Dung! Trang thai lap qua nhieu lan.")
        break
```

```python
import random

# 0 = sach, 1 = ban, 2 = chuong ngai vat

def hien_thi(phong, pos):
    r, c = pos
    print("+---+---+---+")
    for i in range(3):
        hang = "|"
        for j in range(3):
            if (i, j) == (r, c):
                hang += " R |"
            elif phong[i][j] == 0:
                hang += "   |"
            elif phong[i][j] == 1:
                hang += " * |"
            elif phong[i][j] == 2:
                hang += " X |"
        print(hang)

def sensor(phong, pos):
    r, c = pos
    return phong[r][c]

def tim_o_ban_gan_nhat(phong, pos):
    r, c = pos
    tot_nhat = None
    kc_min = float("inf")
    for i in range(3):
        for j in range(3):
            if phong[i][j] == 1:
                kc = abs(i - r) + abs(j - c)
                if kc < kc_min:
                    kc_min = kc
                    tot_nhat = (i, j)
    return tot_nhat

def processor(tinh_trang, pos, phong):
    r, c = pos

    # Dang dung o ban -> hut ngay
    if tinh_trang == 1:
        return "HUT"

    # Tim o ban gan nhat
    dich = tim_o_ban_gan_nhat(phong, pos)
    if dich is None:
        return "DUNG"

    dr, dc = dich[0] - r, dich[1] - c

    # Tinh huong hop le
    huong_doc = None
    huong_ngang = None

    if dr < 0 and r - 1 >= 0 and phong[r-1][c] != 2:
        huong_doc = "LEN"
    elif dr > 0 and r + 1 <= 2 and phong[r+1][c] != 2:
        huong_doc = "XUONG"

    if dc < 0 and c - 1 >= 0 and phong[r][c-1] != 2:
        huong_ngang = "TRAI"
    elif dc > 0 and c + 1 <= 2 and phong[r][c+1] != 2:
        huong_ngang = "PHAI"

    # Uu tien chieu xa hon
    if abs(dr) >= abs(dc):
        if huong_doc:
            return huong_doc
        elif huong_ngang:
            return huong_ngang
    else:
        if huong_ngang:
            return huong_ngang
        elif huong_doc:
            return huong_doc

    # Bi chan ca 2 huong -> di vong qua huong khac con trong
    for h, (nr, nc) in [("LEN",(r-1,c)),("XUONG",(r+1,c)),("TRAI",(r,c-1)),("PHAI",(r,c+1))]:
        if 0 <= nr <= 2 and 0 <= nc <= 2 and phong[nr][nc] != 2:
            return h

    return "DUNG"

def actuator(hanh_dong, pos, phong):
    r, c = pos

    if hanh_dong == "HUT":
        phong[r][c] = 0
        print(f"-> HUT bui tai ({r},{c})")
    elif hanh_dong == "LEN":
        pos = (r - 1, c)
        print(f"-> Di LEN den ({r-1},{c})")
    elif hanh_dong == "XUONG":
        pos = (r + 1, c)
        print(f"-> Di XUONG den ({r+1},{c})")
    elif hanh_dong == "TRAI":
        pos = (r, c - 1)
        print(f"-> Di TRAI den ({r},{c-1})")
    elif hanh_dong == "PHAI":
        pos = (r, c + 1)
        print(f"-> Di PHAI den ({r},{c+1})")
    elif hanh_dong == "DUNG":
        print(f"-> DUNG tai ({r},{c})")

    return pos, phong

def con_ban(phong):
    for hang in phong:
        for o in hang:
            if o == 1:
                return True
    return False

def main():
    print("Nhap phong 3x3 (0=sach, 1=ban, 2=chuong ngai vat)")

    phong = []
    for i in range(3):
        while True:
            raw = input(f"Hang {i+1}: ").split()
            if len(raw) == 3 and all(x in "012" for x in raw):
                phong.append(list(map(int, raw)))
                break
            print("Nhap lai, chi dung 0 1 2")

    while True:
        raw = input("Vi tri robot: ").split()
        if len(raw) == 2:
            r, c = int(raw[0]), int(raw[1])
            if 0 <= r <= 2 and 0 <= c <= 2 and phong[r][c] != 2:
                pos = (r, c)
                break
        print("Vi tri khong hop le, thu lai")

    print()
    hien_thi(phong, pos)

    buoc = 0
    while con_ban(phong):
        buoc += 1
        print(f"--- Buoc {buoc} ---")
        tinh_trang = sensor(phong, pos)
        hanh_dong  = processor(tinh_trang, pos, phong)

        if hanh_dong == "DUNG":
            print("Bi ket, khong the tiep tuc.")
            break

        pos, phong = actuator(hanh_dong, pos, phong)
        hien_thi(phong, pos)

    if not con_ban(phong):
        print(f"Phong da sach sau {buoc} buoc! ===")

if __name__ == "__main__":
    main()
```
